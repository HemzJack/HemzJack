```python
# Task 1 : Mendapatkan daftar perusahaan dengan urutan terbalik
all_companies = [{'name' : namecompany['name']} for namecompany in company_detail_list [::-1]]
# Menampilkan hasil
print(json.dumps(all_companies, indent=2))
```

```python
# Task 2: Mencetak nilai Domain di setiap perusahaan.
def all_domain():
    for company in company_detail_list:
        print(f"{company['name']}: {company['domain']} ({company['country']})")

all_domain()
```

```python
# Task 3: Menampilkan semua karyawan berdasarkan domain perusahaan.
def all_employees():
    domain_employees = {}
    for company in company_detail_list:
        domain = company['domain']
        # Mengambil nama karyawan untuk setiap domain
        domain_employees[domain] = [
            emp['full_name'] for emp in employee_detail_list if emp['company'] == company['name']
        ]
    return domain_employees

# Menampilkan hasil dalam format JSON
print(json.dumps(all_employees(), indent=4))
```

```python
# Task 4: Fungsi untuk mendapatkan karyawan beserta negara perusahaan mereka.
def return_employees():
    employee_country = []
    for emp in employee_detail_list:
        for company in company_detail_list:
            if emp['company'] == company['name']:
                employee_country.append({
                    "full_name": emp['full_name'],
                    "company": emp['company'],
                    "country": company['country']
                })
    return employee_country

print(json.dumps(return_employees(), indent=4))
```

```python
def return_companies():
    company_nationality = []

    # Create a dictionary to store the nationality count for each company
    company_employee_nationality = {}
    for company in company_detail_list:
        company_employee_nationality[company['name']] = {}

    # Count the number of employees per nationality for each company
    for employee in employee_detail_list:
        company_name = employee['company']
        nationality = employee['nationality']
        if nationality in company_employee_nationality[company_name]:
            company_employee_nationality[company_name][nationality] += 1
        else:
            company_employee_nationality[company_name][nationality] = 1

    # Construct the final result
    for company, nationality_count in company_employee_nationality.items():
        company_nationality.append({
            "company": company,
            "employee_nationality": nationality_count
        })

    return company_nationality

# Menampilkan hasil
print(json.dumps(return_companies(), indent=4))
```
